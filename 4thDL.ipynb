# **Learning Rates**

```python
dflist = []

learning_rates = [0.01, 0.05, 0.1, 0.5]

for lr in learning_rates:
    ann = Sequential()
    ann.add(Dense(units=30, activation='relu'))
    ann.add(Dense(units=16, activation='relu'))
    ann.add(Dense(units=9, activation='relu'))
    ann.add(Dense(units=1, activation='linear'))

    opt = tf.keras.optimizers.Adam(learning_rate=lr)
    ann.compile(loss='mean_squared_error',
                  optimizer=opt,
                  metrics=['mae'])
    h = ann.fit(X_train, Y_train, batch_size = 32, validation_data = (X_test,Y_test), epochs = 10)
    
    dflist.append(pd.DataFrame(h.history, index=h.epoch))
```

```python
historydf = pd.concat(dflist, axis=1)
```

```python
historydf
```

```python
# b). 2-class problem

df1 = pd.read_csv('/content/mushroom_dataset.csv')
```

```python
df1.info()
```

```python
df1.drop_duplicates(inplace=True)
```

```python
for i in range(df1.shape[1]):
  print(df1.iloc[:, i].unique())
```

```python
df1.describe()
```

```python
X = pd.get_dummies(df1.iloc[:, :22])
```

```python
X.head(2)
```

```python
y = df1.loc[:, 'mushroom']
```

```python
y = y.map({'EDIBLE': 1, 'POISONOUS': 0})
```

```python
x_train , x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 4)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
ann = Sequential()

ann.add(Dense(units=117, activation='relu'))
ann.add(Dense(units=59, activation='relu'))
ann.add(Dense(units=30, activation='relu'))
ann.add(Dense(units=15, activation='relu'))
ann.add(Dense(units=1, activation='sigmoid'))
```

```python
opt = tf.keras.optimizers.Adam()
loss = tf.keras.losses.BinaryCrossentropy()
```

```python
ann.compile(optimizer = opt, loss = loss, metrics=["accuracy"])
```

```python
ann.fit(x_train, y_train, batch_size=32, epochs=100, validation_data=(x_test, y_test))
```

```python
y_pred = ann.predict(x_test)
```

```python
y_pred = y_pred > 0.5
```

```python
ann.evaluate(x_test, y_test)
```
